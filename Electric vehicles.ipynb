```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Data Loading
data = pd.read_csv('/content/Electric_Vehicle_Population_Data.csv')

# 2. Data Exploration
print("Data Head:")
print(data.head())
print("\nData Info:")
print(data.info())
print("\nData Describe:")
print(data.describe())

# 3. Data Preprocessing
# No preprocessing needed for this example

# 4. Visualization

#
# Create a figure with subplots
fig, axs = plt.subplots(1, 3, figsize=(20, 6))  # 1 row, 3 columns

# Subplot 1: Model Year Distribution
sns.countplot(x='Model Year', data=data, ax=axs[0])
axs[0].set_title('Model Year Distribution')
axs[0].set_xlabel('Model Year')
axs[0].set_ylabel('Count')

# Subplot 2: Electric Range vs. Base MSRP
sns.scatterplot(x='Electric Range', y='Base MSRP', data=data, ax=axs[1])
axs[1].set_title('Electric Range vs. Base MSRP')
axs[1].set_xlabel('Electric Range')
axs[1].set_ylabel('Base MSRP')

# Subplot 3: Make Distribution
sns.countplot(x='Make', data=data, ax=axs[2])
axs[2].set_title('Distribution of Vehicle Makes')
axs[2].set_xlabel('Make')
axs[2].set_ylabel('Count')
axs[2].tick_params(axis='x', rotation=45)  # Rotate x-axis labels for better readability

plt.tight_layout()  # Adjust layout to avoid overlapping titles
plt.show()
```
